### 3. TOTALE CASI COVID-19 PER CONTINENTE

```python
print("| TOTALE CASI COVID-19 PER CONTINENTE |")
def totcasi(covid19_data, cnt1, cnt2, cnt3, cnt4, cnt5, cnt6):
    cnd_1=covid19_data.loc[:, 'continent'] == cnt1
    cnd_2=covid19_data.loc[:, 'continent'] == cnt2
    cnd_3=covid19_data.loc[:, 'continent'] == cnt3
    cnd_4=covid19_data.loc[:, 'continent'] == cnt4
    cnd_5=covid19_data.loc[:, 'continent'] == cnt5
    cnd_6=covid19_data.loc[:, 'continent'] == cnt6
    
    casicovidnw1=covid19_data.loc[cnd_1, 'new_cases'].sum()
    casicovidnw2=covid19_data.loc[cnd_2, 'new_cases'].sum()
    casicovidnw3=covid19_data.loc[cnd_3, 'new_cases'].sum()
    casicovidnw4=covid19_data.loc[cnd_4, 'new_cases'].sum()
    casicovidnw5=covid19_data.loc[cnd_5, 'new_cases'].sum()
    casicovidnw6=covid19_data.loc[cnd_6, 'new_cases'].sum()

    tot_cas={"CONTINENTE":[cnt1, cnt2, cnt3, cnt4, cnt5, cnt6], "CASI TOTALI":[casicovidnw1, casicovidnw2, casicovidnw3, casicovidnw4, casicovidnw5, casicovidnw6]}
        
    return pd.DataFrame(tot_cas)

totcasi(covid19_data, "Africa", "Asia", "Europe", "North America", "Oceania", "South America")
```

### 4. CONFRONTO DATI COVID-19 PER CONTINENTE

```python
print("| CONFRONTO DATI COVID-19 PER CONTINENTE |")
def inf_conf(covid19_data, cnt1, cnt2, cnt3, cnt4, cnt5, cnt6):
    cnd_1=covid19_data.loc[:, 'continent'] == cnt1
    cnd_2=covid19_data.loc[:, 'continent'] == cnt2
    cnd_3=covid19_data.loc[:, 'continent'] == cnt3
    cnd_4=covid19_data.loc[:, 'continent'] == cnt4
    cnd_5=covid19_data.loc[:, 'continent'] == cnt5
    cnd_6=covid19_data.loc[:, 'continent'] == cnt6
    
    casicovidtot1=covid19_data.loc[cnd_1, 'total_cases']
    casicovidtot2=covid19_data.loc[cnd_2, 'total_cases']
    casicovidtot3=covid19_data.loc[cnd_3, 'total_cases']
    casicovidtot4=covid19_data.loc[cnd_4, 'total_cases']
    casicovidtot5=covid19_data.loc[cnd_5, 'total_cases']
    casicovidtot6=covid19_data.loc[cnd_6, 'total_cases']
    
    mincov1=casicovidtot1.min()
    mincov2=casicovidtot2.min()
    mincov3=casicovidtot3.min()
    mincov4=casicovidtot4.min()
    mincov5=casicovidtot5.min()
    mincov6=casicovidtot6.min()
    
    casicovidnw1=covid19_data.loc[cnd_1, 'new_cases']
    casicovidnw2=covid19_data.loc[cnd_2, 'new_cases']
    casicovidnw3=covid19_data.loc[cnd_3, 'new_cases']
    casicovidnw4=covid19_data.loc[cnd_4, 'new_cases']
    casicovidnw5=covid19_data.loc[cnd_5, 'new_cases']
    casicovidnw6=covid19_data.loc[cnd_6, 'new_cases']
    
    maxcov1=casicovidnw1.max()
    maxcov2=casicovidnw2.max()
    maxcov3=casicovidnw3.max()
    maxcov4=casicovidnw4.max()
    maxcov5=casicovidnw5.max()
    maxcov6=casicovidnw6.max()
    
    mediacov1=round(casicovidnw1.mean(),2)
    mediacov2=round(casicovidnw2.mean(),2)
    mediacov3=round(casicovidnw3.mean(),2)
    mediacov4=round(casicovidnw4.mean(),2)
    mediacov5=round(casicovidnw5.mean(),2)
    mediacov6=round(casicovidnw6.mean(),2)
    
    tc=covid19_data.groupby('continent')
        
    percentot1=round(casicovidnw1.sum()*100/tc['new_cases'].sum().sum(),2)
    percentot2=round(casicovidnw2.sum()*100/tc['new_cases'].sum().sum(),2)
    percentot3=round(casicovidnw3.sum()*100/tc['new_cases'].sum().sum(),2)
    percentot4=round(casicovidnw4.sum()*100/tc['new_cases'].sum().sum(),2)
    percentot5=round(casicovidnw5.sum()*100/tc['new_cases'].sum().sum(),2)
    percentot6=round(casicovidnw6.sum()*100/tc['new_cases'].sum().sum(),2)
    
    ccov={"CONTINENTE":[cnt1, cnt2, cnt3, cnt4, cnt5, cnt6], "MINIMO":[mincov1, mincov2, mincov3, mincov4, mincov5, mincov6], "MASSIMO":[maxcov1, maxcov2, maxcov3, maxcov4, maxcov5, maxcov6], 
          "MEDIA CASI":[mediacov1, mediacov2, mediacov3, mediacov4, mediacov5, mediacov6], "% INFEZIONI":[percentot1, percentot2, percentot3, percentot4, percentot5, percentot6]}
    
    return pd.DataFrame(ccov)

inf_conf(covid19_data, "Africa", "Asia", "Europe", "North America", "Oceania", "South America")
```
